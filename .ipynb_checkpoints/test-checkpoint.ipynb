```python
from cv2.ximgproc import *
import cv2
import matplotlib.pyplot as plt
import skimage
import numpy as np
from sklearn import linear_model
import scipy.misc
from sklearn import metrics as mr
import math
```

```python
# step1 对clock1、clock2做导引滤波
img_clock1 = cv2.imread("./images/Clock1.png", 0).astype(np.float32)
img_clock2 = cv2.imread("./images/Clock2.png", 0).astype(np.float32)

guide1 = img_clock1
guide2 = img_clock2

# 进行导向滤波(参数 radius、eps)
Jclock1 = cv2.ximgproc.guidedFilter(
            guide=guide1, src=img_clock1, radius=15, eps=1000, dDepth=-1)
Jclock2 = cv2.ximgproc.guidedFilter(
            guide=guide2, src=img_clock2, radius=15, eps=1000, dDepth=-1)
```

```python
# 绘制图片
plt.figure(figsize=(8,8))
plt.axis('off')

images = [img_clock1, img_clock2, Jclock1, Jclock2]
titles = ['Clock1', 'Clock2', 'Jclock1', 'Jclock2']

plt.subplot(2, 2, 1),plt.imshow(images[0], cmap = plt.cm.gray)  
plt.title(titles[0], fontsize=10)

plt.subplot(2, 2, 2),plt.imshow(images[1], cmap = plt.cm.gray)  
plt.title(titles[1], fontsize=10)

plt.subplot(2, 2, 3),plt.imshow(images[2], cmap = plt.cm.gray)  
plt.title(titles[2], fontsize=10)

plt.subplot(2, 2, 4),plt.imshow(images[3], cmap = plt.cm.gray)  
plt.title(titles[3], fontsize=10)
```

```python
# 保存 Jclock1、Jclock2
cv2.imwrite("./images/Jclock1.png", images[2])
cv2.imwrite("./images/Jclock2.png", images[3])
```

```python
class KSVD(object):
    def __init__(self, n_components=512, max_iter=30, tol=1e-6,
                 n_nonzero_coefs=None):
        """
        稀疏模型Y = DX，Y为样本矩阵，使用KSVD动态更新字典矩阵D和稀疏矩阵X
        :param n_components: 字典所含原子个数（字典的列数）
        :param max_iter: 最大迭代次数
        :param tol: 稀疏表示结果的容差
        :param n_nonzero_coefs: 稀疏度
        """
        self.dictionary = None
        self.sparsecode = None
        self.max_iter = max_iter
        self.tol = tol
        self.n_components = n_components
        self.n_nonzero_coefs = n_nonzero_coefs

    def _initialize(self, y):
        """
        用随机二阶单位范数初始化字典矩阵
        """
        shape=[64,self.n_components]
        #对每一列归一化为L2-norm
        self.dictionary = np.random.random(shape)
        for i in range(shape[1]):
            self.dictionary[:, i]=self.dictionary[:, i]/np.linalg.norm(self.dictionary[:, i])

    def _update_dict(self, y, d, x):
        """
        使用KSVD更新字典的过程
        """
        for i in range(self.n_components):
            index = np.nonzero(x[i, :])[0]
            if len(index) == 0:
                continue
            d[:, i] = 0
            r = (y - np.dot(d, x))[:, index]
            u, s, v = np.linalg.svd(r, full_matrices=False)
            d[:, i] = u[:, 0].T
            x[i, index] = s[0] * v[0, :]
        
        self.sparsecode = x
        return d, x


    def fit(self, y):
        """
        KSVD迭代过程
        """
#         patch_num=int(img.size/64)
        
        #将图像按8*8的块转化列向量，合起来成为64*1024的矩阵
        #img保存原始图像转化的矩阵，y用于保存img减去列均值后的矩阵
#         y=np.zeros((8*8, patch_num))

#         patch_width = int(img.shape[1]/8) # 64
        
#         for patch_index in range(patch_num):
#             #按先行后列，将图片分解成32*32个8*8的小块并装换为列向量
#             r=int(int(patch_index/patch_width)*8)
#             c=int(int(patch_index%patch_width)*8)
#             patch=img[r:r+8, c:c+8].flat
#             y[:, patch_index]=patch

        #字典初始化
        self._initialize(y)
        for i in range(self.max_iter):
            #linear_model.orthogonal_mp 用法详见：
            #http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.orthogonal_mp.html
            x = linear_model.orthogonal_mp(self.dictionary, y, n_nonzero_coefs=self.n_nonzero_coefs)#OMP
            e = np.linalg.norm(y- np.dot(self.dictionary, x))
            print('第%s次迭代，误差为：%s' %(i, e))
            if e < self.tol:
                break
            self._update_dict(y, self.dictionary, x)

        return self.dictionary, self.sparsecode
```

```python
#计算PSNR值
def psnr(A, B):
    if (A==B).all(): return 0
    return 10*np.log10(255*255.0/(((A.astype(np.float)-B)**2).mean()))
# PSBR
def psnr2(img1, img2):
    mse = np.mean( (img1 - img2) ** 2 )
    if mse == 0:
        return 100
    PIXEL_MAX = 255.0
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))

#将8*8块为列向量的矩阵还原为原矩阵
def patch_to_img(patchs, img_rows, img_cols):
    patch_num=patchs.shape[1]
    size=int(img_cols/8)
    img=np.zeros((img_rows, img_cols))
    for i in range(patch_num):
        r=int(int(i/size)*8)
        c=int(int(i%size)*8)
        img[r:r+8, c:c+8]=patchs[:, i].reshape((8, 8))
    return img

#将图像分割为8*8块作为列向量
def img_to_patch(img):
    patchs=np.zeros((8*8, 32*32))
    blocks_r=img.shape[0]/8
    blocks_c=img.shape[1]/8
    patch_num=blocks_r*blocks_c
    for i in range(patchs.shape[1]):
        #按先行后列，将图片分解成32*32个8*8的小块并装换为列向量
        r=int(int(i/blocks_r)*8)
        c=int(int(i%blocks_c)*8)
        patch=img[r:r+8, c:c+8].flat
        patchs[:, i]=patch
    return patchs

# 计算 MI值
def MI(A,B):
    return mr.mutual_info_score(A,B)
```

```python
# 原图
ksvd_clock1 = img_clock1
ksvd_clock2 = img_clock2

# 轮廓图
ksvd_Jclock1 = Jclock1
ksvd_Jclock2 = Jclock2

# 细节图
ksvd_sub1 =cv2.subtract(ksvd_clock1, ksvd_Jclock1)
ksvd_sub2 =cv2.subtract(ksvd_clock2, ksvd_Jclock2)

# 轮廓 patches
ksvd_Jclock1_patches = img_to_patch(ksvd_Jclock1)
ksvd_Jclock2_patches = img_to_patch(ksvd_Jclock2)

# 细节 patches
ksvd_sub1_patches = img_to_patch(ksvd_sub1)
ksvd_sub2_patches = img_to_patch(ksvd_sub2)
```

```python
# 拼接得到 64*2048的y
ksvd_train_y = np.hstack((ksvd_Jclock1_patches, ksvd_Jclock2_patches))
ksvd_train_y_sub = np.hstack((ksvd_sub1_patches, ksvd_sub2_patches))
print('轮廓train', ksvd_train_y.shape)
print('细节train', ksvd_train_y_sub.shape)
```

```python
# 最大迭代次数设为20
ksvd = KSVD(max_iter=20)
print('轮廓图训练：')
ksvd_dictionary, ksvd_sparsecode = ksvd.fit(ksvd_train_y)
print('细节图训练：')
ksvd_dictionary_sub, ksvd_sparsecode_sub = ksvd.fit(ksvd_train_y_sub)
```

```python
# 稀疏系数X1、X2
ksvd_x1 = ksvd_sparsecode[:,0:1024]
ksvd_x2 = ksvd_sparsecode[:,1024:]

ksvd_x1_sub = ksvd_sparsecode_sub[:,0:1024]
ksvd_x2_sub = ksvd_sparsecode_sub[:,1024:]
```

```python
# (X1, X2) => X3
[rows, cols] = ksvd_x1.shape
ksvd_x3 = np.zeros((rows, cols))
ksvd_x3_sub = np.zeros((rows, cols))

for i in range(cols):
    col_x1 = ksvd_x1[:,i]
    col_x2 = ksvd_x2[:,i]
    col_x1_sub = ksvd_x1_sub[:,i]
    col_x2_sub = ksvd_x2_sub[:,i]
    norm1=np.linalg.norm(col_x1, ord=1)
    norm2=np.linalg.norm(col_x2, ord=1)
    norm1_sub=np.linalg.norm(col_x1_sub, ord=1)
    norm2_sub=np.linalg.norm(col_x2_sub, ord=1)
    # 基本图像融合规则
    if(norm1 > norm2):
        ksvd_x3[:,i] = col_x1
    else:
        ksvd_x3[:,i] = col_x2
    # 细节图像融合规则
    if(norm1_sub > norm2_sub):
        ksvd_x3_sub[:,i] = col_x1_sub
    else:
        ksvd_x3_sub[:,i] = col_x2_sub
print(ksvd_x3.shape)
print(ksvd_x3_sub.shape)
```

```python
# 重构基本图片&细节图片
res = ksvd_dictionary.dot(ksvd_x3)
res_sub = ksvd_dictionary_sub.dot(ksvd_x3_sub)

# patches转图片
res_img = patch_to_img(res, 256, 256)
res_sub_img = patch_to_img(res_sub, 256, 256)

# 基本图片
plt.axis('off')
plt.title('res_img')
plt.subplot(1, 2, 1)
plt.imshow(res_img.astype(np.float32), plt.cm.gray)

# 细节图片
plt.axis('off')
plt.title('res_sub_img')
plt.subplot(1, 2, 2)
plt.imshow(res_sub_img.astype(np.float32), plt.cm.gray)

print(psnr(ksvd_Jclock1 ,res_img))
print(psnr(ksvd_Jclock2 ,res_img))
print(psnr(ksvd_sub1 ,res_sub_img))
print(psnr(ksvd_sub2 ,res_sub_img))
```

```python
res_all_img = cv2.add(res_img, res_sub_img)

plt.axis('off')
plt.title('back_img')
plt.imshow(res_all_img.astype(np.float32), plt.cm.gray)
print(psnr(ksvd_clock1 ,res_all_img))
print(psnr(ksvd_clock2 ,res_all_img))
print(psnr2(ksvd_clock1 ,res_all_img))
print(psnr2(ksvd_clock2 ,res_all_img))
```
