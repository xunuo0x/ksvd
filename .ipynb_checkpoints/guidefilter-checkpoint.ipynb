```python
import cv2
import matplotlib.pyplot as plt
import skimage
import numpy as np
```

```python
#计算PSNR值
def psnr(A, B):
    if (A==B).all(): return 0
    return 10*np.log10(255*255.0/(((A.astype(np.float)-B)**2).mean()))
```

```python
img = cv2.imread("./images/cat.bmp",0).astype(np.float32)
print(img)
img = img/255.0
guide = img
print(img)
```

```python
# 进行导向滤波
dst1 = cv2.ximgproc.guidedFilter(
            guide=guide, src=guide, radius=7, eps=100, dDepth=-1)
dst2 = cv2.ximgproc.guidedFilter(
            guide=guide, src=guide, radius=5, eps=1000, dDepth=-1)
dst3 = cv2.ximgproc.guidedFilter(
            guide=guide, src=guide, radius=5, eps=10000, dDepth=-1)
```

```python
# 绘制图片
images = [img,[dst1,dst2,dst3]]
titles =    [
                'Original',
                ['Guided Filter eps=100','Guided Filter eps=200','Guided Filter eps=1000']
            ]

# 绘制原图
plt.figure(figsize=(9,4))
plt.axis('off')

plt.subplot(2, 3, 2),plt.imshow(images[0], cmap = plt.cm.gray)  
plt.title(titles[0], fontsize=10),plt.xticks([]), plt.yticks([])

plt.subplot(2, 3, 4),plt.imshow(images[1][0], cmap = plt.cm.gray)
plt.title(titles[1][0], fontsize=10),plt.xticks([]), plt.yticks([])

plt.subplot(2, 3, 5),plt.imshow(images[1][1], cmap = plt.cm.gray)
plt.title(titles[1][1], fontsize=10),plt.xticks([]), plt.yticks([])

plt.subplot(2, 3, 6),plt.imshow(images[1][2], cmap = plt.cm.gray)
plt.title(titles[1][2], fontsize=10),plt.xticks([]), plt.yticks([])

plt.savefig('1_out.png', transparent=True, dpi=300, pad_inches = 0)
plt.show()
```

```python
sub = cv2.subtract(img, dst1)
plt.axis('off')
plt.title(u"细节图")
plt.imshow(sub.astype(np.float32), plt.cm.gray)

cv2.imwrite('detail_dst1.png', dst1)
```
